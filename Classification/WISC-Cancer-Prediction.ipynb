```python
X = wisc[['radius_worst','concave points_worst']]
y = wisc['diagnosis']
```

```python
X.describe()
```

```python
X_scaled = pd.DataFrame(minmax_scale(X),columns = X.columns)
X_scaled.describe()
```

```python
#140+140+140+140
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=10,test_size=120)
X_train.shape,y_train.shape
```

```python
model = KNeighborsClassifier(n_neighbors=11,weights='distance',p=2)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
y_predict = model.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_predict,y_test)
```
